### **Задание 1.**
Переведите содержимое файла purchase_log.txt в словарь purchases вида: {'1840e0b9d4': 'Продукты', ...}

Пример работы программы при выводе первых двух элементов словаря purchases:
1840e0b9d4 ‘Продукты‘
4e4f90fcfb ‘Электроника‘

```python
from json import loads
```

```python
def from_log_to_dict(filename: str = 'purchase_log.txt') -> dict:
    """
    Функция открывает текстовый файл filename (по умолчанию 'purchase_log.txt') и преобразует содержимое в словарь
    Args:
        - filename: текстовый лог-файл для преобразования;
    Returns:
        - purchases - dict - словарь вида: {'1840e0b9d4': 'Продукты', ...};
        - user_id - str - один из ключей каждого словаря;
        - category - str - один из ключей каждого словаря
    """
    purchases = dict()
    with open(filename, encoding='utf-8') as f:   
        for i, line in enumerate(f):
            dict_ = loads(line.strip())
            
            if i == 0:  
                user_id, category = set(dict_.keys())
                continue

            key1 = dict_.get(user_id, 'Oops!')
            key2 = dict_.get(category, 'Not Found!')

            purchases[key1] = key2

    return purchases, user_id, category
```

```python
purchases, *_ = from_log_to_dict()

i = 0
for key, val in purchases.items():
    i += 1
    print(key, val, sep=' : ')

    if i >= 2:
        break
```

### **Задание 2.**
Для каждого user_id в файле visit_log.csv определите третий столбец с категорией покупки Если покупка была, сам файл visit_log.csv изменять не надо.

Запишите в файл funnel.csv визиты из файла visit_log.csv, в которых были покупки с указанием категории.

Учтите условия для данных:
- содержимое purchase_log.txt помещается в оперативную память компьютера;
- содержимое visit_log.csv — нет, используйте только построчную обработку этого файла

Примеры работы программы при выводе первых трёх строк файла funnel.csv:
user_id,source,category
1840e0b9d4,other,Продукты
4e4f90fcfb,context,Электроника

```python
def insert_third_column(csv: str = 'visit_log.csv', txt: str = 'purchase_log.txt', output: str = 'funnel.csv') -> None:
    """
    Функция, третий столбец с категорией покупки.
    Если покупка была, файл visit_log.csv изменять не надо.
    Args:
        - csv: csv-файл с двумя столбцами: user_id и source;
        - txt: txt-файл со столбцами user_id и category, представлен в виде словарей.
    Returns:
        None
    """
    purchases, user_id, category = from_log_to_dict(filename=txt)

    with open(csv, 'r', encoding='utf-8') as visits, \
         open(output, 'a', newline='', encoding='utf-8') as funnel:
        
        headers = visits.readline().strip().split(';')

        funnel.write(";".join(set(headers + [user_id, category])) + '\n')

        for line in visits:
            line = line.strip().split(';')
            
            user_id, source = line

            if user_id in purchases:
                funnel.write(f"{user_id};{source};{category}\n")
        
    return
```
