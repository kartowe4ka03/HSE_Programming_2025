### **Задание 1.**
Реализуйте функцию sum_distance(from, to), которая суммирует все числа от значения from до величины to включительно.

Примечание. Если пользователь задаст первое число, которое окажется больше второго, просто поменяйте их местами

```python
def sum_distance(fr: int, to: int) -> int:
    """
    Функция суммирует все числа от значения fr до величины to включительно.
    Если первое число больше второго, то они меняются местами.
    Args:
        - fr: Начало суммирования;
        - to: Конец суммирования (включительно).
    Returns:
        - sum(range(fr, to + 1)) - Сумму от fr до to включительно
    """
    if fr > to:
        fr, to = to, fr
    
    return sum(range(fr, to + 1))
```

```python
print(sum_distance(fr=0, to=4) == 10)
print(sum_distance(fr=4, to=0) == 10)
print(sum_distance(fr=15, to=10) == 75)
print(sum_distance(fr=13, to=19) == 112)
```

```python
# Решение задачи со сверточной функцией
anonimous_sum_distance = lambda fr, to: sum(range(fr, to + 1)) if fr < to else sum(range(to, fr + 1))
```

```python
print(anonimous_sum_distance(fr=1, to=100) == 5050)
print(anonimous_sum_distance(fr=10, to=5) == 45)
print(anonimous_sum_distance(fr=12, to=14) == 39)
print(anonimous_sum_distance(fr=79, to=75) == 385)
```

### **Задание 2.**
Реализуйте функцию trim_and_repeat(), которая принимает три параметра:
- Строку;
- offset — число символов, на которое нужно обрезать строку слева;
- repetitions — сколько раз нужно повторить строку перед возвратом
получившейся строки.

Число символов для среза по умолчанию равно 0, а количество повторений — 1.

Функция должна возвращать полученную строку.

```python
def trim_and_repeat(string: str, offset: int = 0, repetitions: int = 1) -> str:
    """
    Функция, обрезающая слева строку string на количество символов offset repetitions раз.
    Args:
        - string: Строка, подвергаемая обрезанию слева;
        - offset: Число символов, на которое нужно обрезать строку string;
        - repetitions: Сколько раз нужно повторить строку перед возвратом получившейся строки.
    Returns:
        Строка, прошедшая все вышеописанные страшные махинации.
    """
    for _ in range(repetitions):
        string = string[offset:]
    
    return string


my_string = "2_Ковалев_МКБ251.ipynb"
```

```python
print(trim_and_repeat(string=my_string))
print(trim_and_repeat(string=my_string, offset=2))
print(trim_and_repeat(string=my_string, offset=2, repetitions=3))
```
